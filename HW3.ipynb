```python
import numpy as np

def load_data(file, pos, neg):
    X, y = [], []
    for line in open(file):
        parts = line.strip().split(',')
        if parts[4] == pos:
            X.append(list(map(float, parts[:4])))
            y.append(1)
        elif parts[4] == neg:
            X.append(list(map(float, parts[:4])))
            y.append(-1)
    return np.array(X), np.array(y)

class Perceptron:
    def __init__(self):
        self.lr = 0.01
        self.epochs = 20

    def train(self, X, y):
        self.w = np.zeros(X.shape[1])
        self.b = 0
        for _ in range(self.epochs):
            for i in range(len(X)):
                if y[i] * (np.dot(self.w, X[i]) + self.b) <= 0:
                    self.w += self.lr * y[i] * X[i]
                    self.b += self.lr * y[i]

    def acc(self, X, y):
        pred = np.sign(np.dot(X, self.w) + self.b)
        return np.mean(pred == y) * 100

def run(c1, c2):
    Xtr, ytr = load_data("A3data/train.data", c1, c2)
    Xte, yte = load_data("A3data/test.data", c1, c2)

    p = Perceptron()
    p.train(Xtr, ytr)

    print(c1, "vs", c2)
    print("Weights:", p.w)
    print("Bias:", p.b)
    print("Train Accuracy:", round(p.acc(Xtr, ytr), 2), "%")
    print("Test Accuracy:", round(p.acc(Xte, yte), 2), "%")
    print("----------------------")

run("class-1", "class-2")
run("class-2", "class-3")
run("class-1", "class-3")
```
